```python
#Cleaning Year Column
ALKO_df["Year"] = pd.to_numeric(ALKO_df["Year"], errors="coerce")
ALKO_df.dropna(subset=["Year"])
ALKO_df = ALKO_df[(ALKO_df["Year"] >= 1921) & (ALKO_df["Year"] <= 2024)]
#cast to integer
ALKO_df["Year"] = ALKO_df["Year"].astype(int)
```

```python
alko = ALKO_df.copy()
for c in alko.columns:
    if c != "Year":
        alko[c] = pd.to_numeric(alko[c].replace("..", pd.NA), errors="coerce")
rename_map = {
    "Sales of lager equivalents, total (1000 litres) (alcohol 4,6 vol.)": ("beer_litres", 1000),
    "Sales of wine, total (1000 litres)": ("wine_litres", 1000),
    "Sales of spirits, total (1000 litres)": ("spirits_litres", 1000),
    "Sales of alcohol, total, pure alcohol (1000 litres)": ("alco_total_pure_litres", 100),
    "Sales of beer, pure alcohol (1000 litres)": ("beer_pure_liters", 1000),
    "Sales of wine, pure alcohol (1000 litres)": ("wine_pure_liters", 1000),
    "Sales of spirits, pure alcohol (1000 litres)": ("spirits_pure_liters", 1000),
}

# Create new columns programmatically
for old, (new, factor) in rename_map.items():
    alko[new] = alko[old] * factor

# Drop old columns in one go
alko = alko.drop(columns=list(rename_map.keys()))
```

```python
PRIS_df = pd.read_csv(PRIS_df_path, header=1)
```

```python
# First, name the columns temporarily
cols = PRIS_df.columns.tolist()
rename_map = {
    cols[0]: "raw_category",
      cols[1]: "period",
    cols[2]: "index",
    cols[3]: "pct_change_mom",
    cols[4]: "pct_change_yoy",
}
PRIS_df = PRIS_df.rename(columns=rename_map)
```

```python

# 3) Identify category header rows (where a new category label appears in the first column)
#    These rows typically have period/index columns as NaN.
is_category_row = PRIS_df["raw_category"].notna() & PRIS_df["period"].isna()
PRIS_df.loc[is_category_row, "category"] = PRIS_df.loc[is_category_row, "raw_category"]
# 4) Forward-fill the category so every data row knows which category it belongs to
PRIS_df["category"] = PRIS_df["category"].ffill()


# 5) Keep only proper data rows — where 'period' looks like YYYYMmm (e.g., 2001M01)
period_mask = PRIS_df["period"].astype(str).str.fullmatch(r"\d{4}M\d{2}")
PRIS_df = PRIS_df[period_mask].copy()

# 6) Parse 'period' to a real date (month start)
#    Convert '2001M01' -> '2001-01-01'
PRIS_df["date"] = pd.to_datetime(
    PRIS_df["period"].str.replace("M", "-", regex=False) + "-01",
    format="%Y-%m-%d",
    errors="coerce",
)

# 7) Clean numeric columns
for col in ["index", "pct_change_mom", "pct_change_yoy"]:
    PRIS_df[col] = pd.to_numeric(PRIS_df[col], errors="coerce")
```

```python
# 0) Start from your `data` right after creating the 'date' column
# Ensure tidy strings & numeric types
PRIS_df["Category"] = PRIS_df["category"].str.strip()
print(PRIS_df["Category"].unique())
for c in ["index", "pct_change_mom", "pct_change_yoy"]:
    PRIS_df[c] = pd.to_numeric(PRIS_df[c], errors="coerce")

# 1) Sort + set MultiIndex
PRIS_df = PRIS_df.sort_values(["Category", "date"]).set_index(["Category", "date"])

# 2) Drop exact duplicates (same category+date)
PRIS_df = PRIS_df[~PRIS_df.index.duplicated(keep="last")]

# 3) Fill MoM / YoY from the index (in percentage points)
def fill_changes_from_index(df):
    # df is one category (indexed by date)
    mom = df["index"].pct_change(1) * 100
    yoy = df["index"].pct_change(12) * 100
    df["pct_change_mom"] = df["pct_change_mom"].fillna(mom)
    df["pct_change_yoy"] = df["pct_change_yoy"].fillna(yoy)
    return df

PRIS_df = PRIS_df.groupby(level=0, group_keys=False).apply(fill_changes_from_index)

# 4) Quick gap check (monthly frequency): list missing months per category
def find_missing_months(df):
    # df is one category (indexed by date)
    idx = df.index.get_level_values("date")
    full = pd.date_range(idx.min(), idx.max(), freq="MS")
    missing = full.difference(idx)
    return pd.Series({"missing_months": list(missing)})

gap_report = (
    PRIS_df.reset_index()
        .groupby("Category")
        .apply(lambda g: find_missing_months(g.set_index("date")))
        .reset_index()
)

# Optional: if you want to forward/backfill pct changes (not index!) for tiny gaps:
# data[["pct_change_mom", "pct_change_yoy"]] = (
#     data.groupby(level=0)[["pct_change_mom", "pct_change_yoy"]]
#         .apply(lambda g: g.sort_index().ffill().bfill())
# )

# Done: tidy table
PRIS_df = PRIS_df[["index", "pct_change_mom", "pct_change_yoy"]]
```
