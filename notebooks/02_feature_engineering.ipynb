```python
import numpy as np
import pandas as pd
```

```python
features = pd.read_csv("../data/processed/joined_features.csv")
```

```python
features = features.sort_values("Year").reset_index(drop=True)
```

```python
# assert (features["beer_litres"] >= 0).all()
# assert (features["Population"] > 0).all()
```

```python
features["Year"] = features["Year"].astype(int)
for col in ["beer_litres", "Population", "cpi_index"]:
    features[col] = pd.to_numeric(features[col], errors="coerce")
```

```python
# per-capita + growth
features["beer_lpc"] = features["beer_litres"] / features["Population"]         # litres per capita
features["beer_lpc_yoy"] = features["beer_lpc"].pct_change()
features["beer_litres_yoy"] = features["beer_litres"].pct_change()
```

```python
features["cpi_yoy"] = features["cpi_index"].pct_change()
```

```python
features
```

```python
#cap for 2015 and above:
features = features[features["Year"]>= 2015]
```

```python
final_cols = [
    "Year",
    "beer_litres",
    "Population",
    "cpi_index",
    "beer_lpc",
    "beer_lpc_yoy",
    "beer_litres_yoy",
    "cpi_yoy"
]

features[final_cols].to_csv("../data/processed/final_features.csv", index=False)
```
